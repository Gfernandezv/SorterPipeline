## Variables

las variables que definen una probe se almacenan en una tabla de excel, la que se referencia en la variable "excel_probefile" con el siguiente formato:  

| info  | Dev | Chan | TTR | Probe |
| :---:   | :---:  | :---:  | :---:  | :---:  |
| informacion del canal  | numero de canal (partiendo en 0) | Canal LAN (partiendo en 1) | Tetrodo (partiendo en 1) | Probe (partiendo en 0) |
|  Ej: "CA1v" |  Ej: "0"  |  Ej: "1" |  Ej: "1" |  Ej: "0" |

 La probe se crea en 4 grupos, cada electrodo se separa en **micrones** y se definen en las variables **"electrode_distance"** y cada grupo en **"group_distance"**  
 El archivo de probe se guarda como **"probe_name**_probe.json" dentro de la carpeta "probes"

```python
probe_name = 'test'
electrode_distance = 20
group_distance = 1000
file_path = r'C:\Users\Labcn\OneDrive - Universidad Cat√≥lica de Chile\sorting_anillo\probes\anillo_probe.xlsx'  # Reemplaza con la ruta a tu archivo Excel
```

```python
# Extraer la columna 'Dev' para usarla en set_global_device_channel_indices
df_excel = pd.read_excel(file_path)
device_channel_indices = df_excel['Dev'].values

# Crear matrices de posiciones (como antes)
n = 4
all_positions = []

for j in range(4):
    positions = np.zeros((n, 2))
    for i in range(n):
        x = j * group_distance  # Avanzar x en saltos de 100
        y = i * electrode_distance   # Avanzar y en saltos de 30
        positions[i] = [x, y]
    all_positions.append(positions)

# Crear un ProbeGroup
probe_group = pi.ProbeGroup()

# Crear y a√±adir probes al ProbeGroup
for positions in all_positions:
    probe = pi.Probe(ndim=2, si_units='um')
    probe.set_contacts(positions=positions, shapes='circle', shape_params={'radius': 5})
    probe.create_auto_shape(probe_type='tip')
    probe_group.add_probe(probe)
    
probe_group.set_global_device_channel_indices(device_channel_indices)
# Graficar el ProbeGroup
plot_probegroup (probe_group)
write_probeinterface(file=("probes/" + probe_name + "_probe.json"), probe_or_probegroup=probe_group)

df = probe_group.to_dataframe()
print(df)
```

```python
import numpy as np
import scipy.io

# Generar la disposici√≥n de los canales
n_groups = 4
n_channels_per_group = 4
n_total_channels = n_groups * n_channels_per_group

# Separaci√≥n de canales dentro de un grupo y entre grupos
channel_spacing = 20  # Separaci√≥n de 20um entre canales dentro de un grupo
group_spacing = 1000    # Separaci√≥n de 50um entre grupos

# Inicializar arrays para las coordenadas
xc = []
yc = []

# Generar las coordenadas
for group_idx in range(n_groups):
    # La posici√≥n en y aumenta para cada grupo, separados por group_spacing
    y_base = group_idx * group_spacing
    for channel_idx in range(n_channels_per_group):
        # Los canales dentro de cada grupo est√°n alineados en x, separados por channel_spacing
        xc.append(channel_idx * channel_spacing)
        yc.append(y_base)

# Convertir a arrays de numpy
xc = np.array(xc)
yc = np.array(yc)

# √çndices de los canales
chanMap = np.arange(1, n_total_channels + 1)

# Todos los canales pertenecen al mismo shank (puedes ajustarlo si es necesario)
kcoords = np.zeros(n_total_channels)

# N√∫mero total de canales
n_chan = n_total_channels

# Crear un diccionario con los datos de la probe
probe_data = {
    'chanMap': chanMap,
    'xc': xc,
    'yc': yc,
    'kcoords': kcoords,
    'n_chan': n_chan
}

from kilosort.io import save_probe

save_probe(probe_data, 'probes/probe_kilosort.json')

print("Archivo kilosort_probe_16ch.mat generado con √©xito.")
```
